```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error


```

```python

df = pd.read_csv('/kaggle/input/login2024/Train.csv')  


print(df.head())

```

```python

X = df['ENTITY_DESCRIPTION']  
y = df['ENTITY_LENGTH']       

vectorizer = TfidfVectorizer(max_features=100)
X_tfidf = vectorizer.fit_transform(X)


X_train, X_test, y_train, y_test = train_test_split(X_tfidf, y, test_size=0.2, random_state=42)
```

```python

model = LinearRegression()
model.fit(X_train, y_train)


first_100_descriptions = df['ENTITY_DESCRIPTION'][:100]
first_100_tfidf = vectorizer.transform(first_100_descriptions)
predicted_lengths = model.predict(first_100_tfidf)


for desc, length in zip(first_100_descriptions, predicted_lengths):
    print(f"Description: {desc}")
    print(f"Predicted Length: {length:.2f} feet\n")
```
