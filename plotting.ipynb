```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python

df = pd.read_csv('/kaggle/input/dataset2/Train.csv')

df_subset = df.head(200)

grouped_data = df_subset.groupby('CATEGORY_ID')['ENTITY_LENGTH'].sum().reset_index()



                 
```

```python

plt.figure(figsize=(10,6))


plt.bar(grouped_data['CATEGORY_ID'], grouped_data['ENTITY_LENGTH'], color='skyblue', width=100)



```

```python
plt.xlabel('Category ID')


plt.ylabel('Total Length (Summed)')


plt.title('Summed Length by Category ID')
```

```python
plt.show()
```
